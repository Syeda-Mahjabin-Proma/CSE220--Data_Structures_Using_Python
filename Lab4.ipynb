```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
    
```

```python
class DoublyList:
  
    def __init__(self, a):
    # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
        self.head = Node(None, None, None)
        temp = self.head
        for i in range(len(a)):
            new_node = Node(a[i], None, None)
            temp.next = new_node
            new_node.prev = temp
            temp = temp.next
        temp.next = self.head
        self.head.prev = temp
    
    # Counts the number of Nodes in the list and return the number
    def countNode(self):
        temp = self.head.next
        count = 0
        while temp != self.head:
            count += 1
            temp = temp.next
        return count
 
    # prints the elements in the list
    def forwardprint(self):
        temp = self.head.next
        while temp != self.head:
            if temp == self.head.prev:
                print(temp.element)
            else:
                print(temp.element, end=", ")
            temp = temp.next
        
    # prints the elements in the list backward
    def backwardprint(self):
        temp = self.head.prev
        while temp != self.head:
            if temp is self.head.next:
                print(temp.element)
            else:
                print(temp.element, end=", ")
            temp = temp.prev

    # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        count_node = self.countNode()
        # print(count_node)
        if idx>= 0 or idx < count_node:
            temp = self.head.next
            c = 0
            while temp is not self.head:
                c += 1
                temp = temp.next
                if c == idx:
                    return temp
        return
     
    # returns the index of the containing the given element. if the element does not exist in the List, return -1.
    def indexOf(self, elem):
        temp = self.head.next
        count = 0
        while temp is not self.head:
            if temp.element is elem:
                return count
            temp = temp.next
            count += 1
        return -1

    # inserts containing the given element at the given index Check validity of index. 
    def insert(self, elem, idx):
        temp = self.head.next
        count = 0
        new_node = Node(elem, None, None)
        count_node = self.countNode()
        if idx < 0 or idx > count_node:
            print('Invalid Index')
        else:         
            while count < idx:
                temp = temp.next
                count += 1
            p = temp.prev
            new_node.next = temp
            temp.prev = new_node
            p.next = new_node
            new_node.prev = p

    # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
    def remove(self, idx):
        temp = self.head.next
        count = 0
        count_node = self.countNode()
        if idx < 0 or idx > count_node-1:
            return
        while temp != self.head:
            if count == idx:
                temp.prev.next = temp.next
                temp.next.prev = temp.prev
                temp.next = None
                temp.prev = None
                return str(temp.element)
            temp = temp.next
            count += 1

```
