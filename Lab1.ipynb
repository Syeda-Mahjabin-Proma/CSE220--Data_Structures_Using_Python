```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
    for x in range(k):
        for y in range(len(source)-1):
            source[y] = source[y+1]
        source[len(source)-1] = 0
    return source


# Test 02: Rotate Left k cell
def rotate_left(source, k):
    for x in range(k):
        temp = source[0]
        for y in range(len(source)-1):
            source[y] = source[y+1]
        source[len(source)-1] = temp
    return source


# Test 03: Shift Right k cell
def shift_right(source, k):
    for x in range(k):
        i = 0
        for y in range(len(source)-1):
            source[len(source)-1-i]= source[len(source)-1-i-1]
            i+=1
        source[0]=0
    return source
        




# Test 04: Rotate Right k cell
def rotate_right(source, k):
    for x in range(k):
        i = 0
        temp = source[len(source)-1]
        for y in range(len(source)-1):
            source[len(source)-1-i]= source[len(source)-1-i-1]
            i+=1
        source[0]=temp
    return source



# Test 05: Remove an element from an array
def remove(source, idx):
    if idx>0 and idx<len(source):
        for x in range(idx + 1, len(source)):
            source[x - 1] = source[x]
        source[len(source) - 1 - 1] = 0
    return source

# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
    for x in range(len(source)):
        if source[x] == 2:
            source[x] = 0
    for y in range(len(source)-1):
        if source[y] == 0:
            for z in range(y, len(source)-1):
                source[z] = source[z+1]
            source[len(source)-1] = 0
    return source

# Test 07: Splitting an Array
def split_array(array):
    result = False
    for i in range(len(array) - 1):
        x = 0
        y = 0
        for j in range(len(array)):
            if j < i:
                x += array[j]
            else:
                y += array[j]
        if x == y:
            result = True
            break

    return result




# Test 08: Max Bunch Count
def max_bunch(a):
    temp = 1
    count = 0
    for i in range(len(a) - 1):
        if a[i] == a[i + 1]:
            temp += 1
            if temp > count:
                count = temp

        else:
            temp = 1
    return count
```

```python
def mean_val(arr):
    sum_1 = 0
    for x in range(len(arr)):
        sum_1 += arr[x]
    mean = sum_1 / len(arr)
    # print(mean)
    return mean


def SD(arr):
    mean = mean_val(arr)
    draft_sum = 0
    for x in range(len(arr)):
        draft_sum += ((arr[x] - mean) ** 2)
    var = draft_sum / (len(arr) - 1)
    sd = var ** 0.5
    # print(sd)
    return sd


def new_array(arr, k):
    new_arr = [0] * 2
    sd = SD(arr)
    mean = mean_val(arr)
    a = mean - (sd * k)
    b = mean + (sd * k)
    for x in range(len(arr)):
        if arr[x] < a:
            new_arr[0] = arr[x]
        if arr[x] > b:
            new_arr[1] = arr[x]
    return new_arr


a = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
print(mean_val(a))
print(SD(a))
print(new_array(a, 1.5))
```
