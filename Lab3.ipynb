```python
# Run This cell Second
class Node:
    def __init__(self, e, n):
        self.element = e
        self.next = n
```

```python
# To you coding in this cell
class LinkedList:
  
    def __init__(self, a):
        if type(a) == list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                n = Node(a[i], None)
                tail.next = n
                tail = tail.next
        else:
            self.head = a
            tail = self.head

    def traverseList(self):
        s = ''
        temp = self.head
        while temp != None:
            if temp.next != None:
                s += str(temp.element) + " "
            else:
                s += str(temp.element)
            temp = temp.next
        return s
    
    # Count the number of nodes in the list and return the total number
    def countNode(self):
        count = 0
        temp = self.head
        while temp is not None:
            count+=1
            temp = temp.next
        # print(count)
        return count

    
    # returns the reference of the Node at the given index. For invalid index return None.
    def nodeAt(self, idx):
        temp = self.head
        count = 0
        temp_2 = 0
        new_none_node = Node(None, None)
        while temp is not None:
            if count == idx:
                temp_2 = temp
            count += 1
            temp = temp.next
        if temp_2 == 0:
            temp_2 = new_none_node
        return temp_2
    
    # returns the element of the Node at the given index. For invalid idx return None.
    def get(self, idx):
        node_count = self.countNode()
        if idx<0 or idx> node_count-1:
            return None
        count = 0
        temp = self.head
        while count != idx:
            count+=1
            temp = temp.next
            if count == idx:
                # print(temp.element)
                return temp.element
    
    # updates the element of the Node at the given index. 
    # Returns the old element that was replaced. For invalid index return None.
    # parameter: index, element
    def set(self, idx, elem):
        count = self.countNode()
        if idx<0 or idx> count-1:
            return None
        
        temp = self.head
        count_2 = 0
        while count_2 != idx:
            count_2+=1
            temp = temp.next
            if count_2 == idx:
                x = temp.element
                temp.element = elem
                # print(x)
                return x

    # returns the index of the Node containing the given element.
    # if the element does not exist in the List, return -1.
    def indexOf(self, elem):
        node_count = self.countNode()
        count = 0
        while count!= node_count:
            temp = self.head
            while temp!= None:
                if temp.element == elem:
                    return count
                else:
                    temp = temp.next
                    count+=1
        return -1
                    
    
    # returns true if the element exists in the List, return false otherwise.
    def contains(self, elem):
        contain = False
        temp = self.head
        while temp!= None:
            if temp.element == elem:
                return True
            else:
                temp = temp.next
        return contain

    # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
    def copyList(self):
        temp = self.head
        new_head = None
        new_temp = None
        while temp != None:
            if new_head == None:
                new_head = Node(temp.element, temp.next)
                new_temp = new_head
            else:
                temp = Node(temp.element, temp.next)
                new_temp.next = temp
                new_temp = new_temp.next
            temp = temp.next
        return new_head


    # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
    def reverseList(self):
        temp = self.head
        new_head = None
        while temp is not None:
            temp_1 = temp.next
            temp.next = new_head
            new_head = temp
            temp = temp_1
        return new_head
    
    # inserts Node containing the given element at the given index
    # Check validity of index. If invalid then print "Invalid Index"
    def insert(self, elem, idx):
        count_node = self.countNode()
        if idx < 0 or idx > count_node:
            print("Invalid Index")
            return "Invalid Index"
        new_node = Node(elem, None)
        if idx == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            temp = self.nodeAt(idx - 1)
            new_node.next = temp.next
            temp.next = new_node

    # removes Node at the given index. returns element of the removed node.
    # Check validity of index. return None if index is invalid.
    def remove(self, idx):
        remove_elem = (self.nodeAt(idx)).element
        remove_node = self.nodeAt(idx)
        if idx == 0:
            self.head = self.head.next
        else:
            temp = self.nodeAt(idx - 1)
            temp.next = remove_node.next

        return remove_elem
    
    # Rotates the list to the left by 1 position.
    def rotateLeft(self):
        temp = self.head
        self.head = self.head.next
        head = self.head
        temp_1 = None
        while head is not None:
            temp_1 = head
            head = head.next
        temp_1.next = temp
        temp.next = None
        return temp
    
    
    # Rotates the list to the right by 1 position.
    def rotateRight(self):
        temp = self.head
        last_node = self.nodeAt(self.countNode()-1)
        before_last_node = self.nodeAt(self.countNode()-2)
        last_node.next = temp
        self.head = last_node
        before_last_node.next = None
        
```
